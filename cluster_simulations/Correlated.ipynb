```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm




%matplotlib inline

font = {'family' : 'serif',
        'size'   : 16}
text = {'usetex' : True}

plt.rc('font', **font)
plt.rc('text',**text)
plt.rc('axes', titlesize=18)
```

```python
file_name = 'correlated0_summary.npz'
with np.load(file_name) as file:
    data = file['data'].item()
```

```python
data.keys()
```

```python
D_vec = data['D_vec']
c_vec = data['c_vec']
gamma = data['gamma']
M = data['M']
epoch_timescale = 40
logN = data['logN']
rep_num = data['rep_num']
mu = data['mu']



num_alive_array = data['num_alive_array']

for ind0 in range(len(D_vec)):

    print('D = {}, \gamma = {}, M = {:.1f}, logN = {} M, epoch_time = {}, mu = {}'.format(D_vec[ind0],gamma,M,logN/M,epoch_timescale,mu))

    plt.figure(figsize=(15,4))
    for ind2 in range(rep_num):
        plt.subplot(1,rep_num,ind2+1)
        for ind1 in range(len(c_vec)):
            D = int(D_vec[ind0])
            num_alive = num_alive_array[ind0,ind1,ind2]
            num_alive[num_alive==0] = np.nan
            plt.plot(mu*np.arange(len(num_alive)),num_alive,label='c={}'.format(c_vec[ind1]))

        if ind2==0:
            plt.legend(prop={'size':12})
            plt.ylabel('Number of types')
            plt.xlabel('Types added')
        plt.title('rep = {}'.format(ind2))

    plt.show()
        
# ind1 = 1
# print(r'\gamma = {}, M = {:.1f}, logN = {} M, epoch_time = {}'.format(g_vec[ind1],M,logN/M,epoch_timescale))

# plt.figure(figsize=(15,4))
# for ind2 in range(rep_num):
#     plt.subplot(1,rep_num,ind2+1)
#     for ind0 in range(len(D_vec)):
#         D = int(D_vec[ind0])
#         num_alive = num_alive_array[ind0,ind1,ind2]
#         num_alive[num_alive==0] = np.nan
#         plt.plot(mu*np.arange(len(num_alive)),num_alive,label='D={}'.format(D))
        
#     if ind2==0:
#         plt.legend(prop={'size':12})
#         plt.ylabel('Number of types')
#         plt.xlabel('Types added')
#     plt.title('rep = {}'.format(ind2))
    
# plt.show()
```
