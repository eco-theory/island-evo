```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm




%matplotlib inline

font = {'family' : 'serif',
        'size'   : 16}
text = {'usetex' : True}

plt.rc('font', **font)
plt.rc('text',**text)
plt.rc('axes', titlesize=18)
```

```python
file_name = 'D_scaling0_summary.npz'
with np.load(file_name) as file:
    data = file['data'].item()
```

```python
D_vec = data['D_vec']
g_vec = data['g_vec']
M = data['M']
epoch_timescale = 40
logN = data['logN']
rep_num = data['rep_num']
mu = data['mu']



num_alive_array = data['num_alive_array']

ind1 = 0

print('\gamma = {}, M = {:.1f}, logN = {} M, epoch_time = {}'.format(g_vec[ind1],M,logN/M,epoch_timescale))

plt.figure(figsize=(15,4))
for ind2 in range(rep_num):
    plt.subplot(1,rep_num,ind2+1)
    for ind0 in range(len(D_vec)):
        D = int(D_vec[ind0])
        num_alive = num_alive_array[ind0,ind1,ind2]
        num_alive[num_alive==0] = np.nan
        plt.plot(mu*np.arange(len(num_alive)),num_alive,label='D={}'.format(D))
        
    if ind2==0:
        plt.legend(prop={'size':12})
        plt.ylabel('Number of types')
        plt.xlabel('Types added')
    plt.title('rep = {}'.format(ind2))
        
plt.show()
        
ind1 = 1
print(r'\gamma = {}, M = {:.1f}, logN = {} M, epoch_time = {}'.format(g_vec[ind1],M,logN/M,epoch_timescale))

plt.figure(figsize=(15,4))
for ind2 in range(rep_num):
    plt.subplot(1,rep_num,ind2+1)
    for ind0 in range(len(D_vec)):
        D = int(D_vec[ind0])
        num_alive = num_alive_array[ind0,ind1,ind2]
        num_alive[num_alive==0] = np.nan
        plt.plot(mu*np.arange(len(num_alive)),num_alive,label='D={}'.format(D))
        
    if ind2==0:
        plt.legend(prop={'size':12})
        plt.ylabel('Number of types')
        plt.xlabel('Types added')
    plt.title('rep = {}'.format(ind2))
    
plt.show()
```

```python
file_name = 'D_scaling0_D3_g1_rep2.npz'
with np.load(file_name) as file:
    data = file['class_obj'].item()
```

```python
dt = data['dt_list'][-1]
t_max = dt*np.shape(data['n_traj_f'])[1]
tvec = np.arange(0,t_max,dt)
mu = data['mu']

plt.figure()

n_traj = data['n_traj_f']
K0 = n_traj.shape[0]

for ii in range(K0-mu):
    plt.plot(tvec,n_traj[ii,:]);
    
for ii in range(K0-mu,K0,1):
    plt.plot(tvec,n_traj[ii,:],'--');



plt.yscale('log')
plt.hlines(y = np.exp(-data['M']),xmin=-0.04*t_max,
           xmax=1.04*t_max,
           linestyle='--',color='k',label='$M$')
# plt.legend()
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Dynamics final evolutionary point',y=1.03)
```
